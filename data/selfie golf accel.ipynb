```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def go(f):
    with open(f, 'r') as fh:
        data = fh.read().strip().split('\n')
    gyro = list()
    accel = list()
    for d in data:
        s = d.split(" ")
        if s[0] == 'accel':
            accel.append(list(map(float, s[1:])))
        else:
            gyro.append(list(map(float, s[1:])))
    
    plt.plot(accel)
    plt.show()
    plt.plot(gyro)
    plt.show()
    return np.array(accel),np.array(gyro)
```

```python
accel, gyro = go('data.1477789475.948549')
```

```python
accel, gyro = go('data.1477787571.0544')
```

```python
accel, gyro = go('data.1477787537.9728587')


```

```python
from scipy.ndimage import filters
dt = .02

r = 1.2
xpp = accel[:-1, 1]
zpp = accel[:-1,2]
Op = gyro[:, 0]
plt.plot(dt * np.cumsum(Op))

plt.plot(np.arcsin(filters.gaussian_filter1d((xpp -  np.gradient(Op) * r/dt)/9.8, 6)))


#plt.plot(np.arccos((zpp - Op**2 * r)/9.8))
plt.show()
```

```python
xs = []
x = 0
weight = .01
unweight = 1 - weight
Oprev = 0
for O, xp in zip(Op, xpp):
    x += dt * O
    grad = O - Oprev
    Oprev = O
    guess = np.arcsin((xp -  grad * r/dt)/9.8)
    if not(np.isnan(guess)):
        x = weight * guess + unweight * x
    xs += [x]
```

```python
plt.plot(np.array(xs))
plt.plot(dt * np.cumsum(Op))
plt.show()
```
