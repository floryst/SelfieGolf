```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('/dev/shm/data.1477787537.9728587', 'r') as fh:
    data1 = fh.read().strip().split('\n')
with open('/dev/shm/data.1477787571.0544', 'r') as fh:
    data2 = fh.read().strip().split('\n')
with open('/dev/shm/data.1477787590.2102954', 'r') as fh:
    data3 = fh.read().strip().split('\n')

gyro1 = list()
accel1 = list()
for d in data1:
    s = d.split(' ')
    if s[0] == 'accel':
        accel1.append(list(map(float, s[1:])))
    else:
        gyro1.append(list(map(float, s[1:])))
    
gyro2 = list()
accel2 = list()
for d in data2:
    s = d.split(' ')
    if s[0] == 'accel':
        accel2.append(list(map(float, s[1:])))
    else:
        gyro2.append(list(map(float, s[1:])))

gyro3 = list()
accel3 = list()
for d in data3:
    s = d.split(' ')
    if s[0] == 'accel':
        accel3.append(list(map(float, s[1:])))
    else:
        gyro3.append(list(map(float, s[1:])))
```

```python
plt.plot(accel1)
plt.show()
plt.plot(accel2)
plt.show()
plt.plot(accel3)
plt.show()

plt.plot(gyro1)
plt.show()
plt.plot(gyro2)
plt.show()
plt.plot(gyro3)
plt.show()
```
