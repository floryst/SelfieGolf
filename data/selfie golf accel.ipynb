```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def go(f):
    with open(f, 'r') as fh:
        data = fh.read().strip().split('\n')
    gyro = list()
    accel = list()
    for d in data:
        s = d.split(" ")
        if s[0] == 'accel':
            accel.append(list(map(float, s[1:])))
        else:
            gyro.append(list(map(float, s[1:])))
    
    plt.plot(accel)
    plt.show()
    plt.plot(gyro)
    plt.show()
    return accel,gyro
```
